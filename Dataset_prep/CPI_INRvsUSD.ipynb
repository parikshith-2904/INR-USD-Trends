```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
ind_cpi = pd.read_csv(r"Datasets\CPI_IND.csv")
ind_cpi
```

```python
ind_cpi = ind_cpi[["TIME_PERIOD:Period","OBS_VALUE:Value"]]
ind_cpi
```

```python
usa_cpi = pd.read_csv(r"Datasets\CPI_USA.csv")
usa_cpi
```

```python
usa_cpi = usa_cpi[["TIME_PERIOD:Period","OBS_VALUE:Value"]]
usa_cpi
```

```python
usa_cpi = usa_cpi.rename(columns={"OBS_VALUE:Value":"USA_CPI"})
ind_cpi = ind_cpi.rename(columns={"OBS_VALUE:Value":"IND_CPI"})
ind_cpi
```

```python
usa_cpi
```

```python
cpi_data = ind_cpi.merge(usa_cpi, how='inner', on="TIME_PERIOD:Period")
cpi_data
```

```python
cpi_data = cpi_data.rename(columns={"TIME_PERIOD:Period":"Date"})
```
