```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ind_cgdp = pd.read_csv(r"Datasets\credit_to_gdp_ind_quarterly.csv")
usa_cgdp = pd.read_csv(r"Datasets\credit_to_gdp_usa_quarterly.csv")
```

```python
usa_cgdp
```

```python
usa_cgdp = usa_cgdp.rename(columns={"TIME_PERIOD:Period":"Date", "OBS_VALUE:Value":"USA_Credit_to_GDP_Ratios"})
usa_cgdp = usa_cgdp[["Date", "USA_Credit_to_GDP_Ratios"]]
usa_cgdp
```

```python
credit_to_gdp = ind_cgdp.merge(usa_cgdp, on="Date", how="inner")
credit_to_gdp
```
