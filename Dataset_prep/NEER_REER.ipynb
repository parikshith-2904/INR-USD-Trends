```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ind_eer = pd.read_csv(r"Datasets\IND_EER.csv")
ind_eer
```

```python
ind_neer = ind_eer[ind_eer["EER_TYPE:Type"] == "N:Nominal"]
ind_neer
```

```python
ind_neer = ind_neer.rename(columns={"TIME_PERIOD:Period":"Date" ,"OBS_VALUE:Value":"IND_NEER"})
ind_neer = ind_neer[["Date", "IND_NEER"]]
ind_neer
```

```python
ind_reer = ind_eer[ind_eer["EER_TYPE:Type"] == "R:Real"]
ind_reer
```

```python
ind_reer = ind_reer.rename(columns={"TIME_PERIOD:Period":"Date" ,"OBS_VALUE:Value":"IND_REER"})
ind_reer = ind_reer[["Date", "IND_REER"]]
ind_reer
```

```python
ind_eer = ind_neer.merge(ind_reer, on="Date", how="inner")
ind_eer
```
